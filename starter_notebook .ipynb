````python
!pip install hands-on-ai gradio pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import gradio as gr
from hands_on_ai.chat import get_response


# -------------------------
# Sample Transaction Dataset
# -------------------------

sample_transactions = {
    'Date': [
        '2024-09-01', '2024-09-02', '2024-09-03', '2024-09-04',
        '2024-09-05', '2024-09-06', '2024-09-07', '2024-09-08',
        '2024-09-09', '2024-09-10'
    ],
    'Amount': [
        '$72.40', '$18.50', '$120.00', '$6.80', '-$20.00',
        '$14.99', '$42.30', '$9.50', '$65.00', '$28.75'
    ],
    'Category': [
        'Food Delivery', 'Fuel', 'Shopping', 'Coffee', 'Refund',
        'Subscription', 'Groceries', 'Transport', 'Entertainment', 'Dining Out'
    ],
    'Description': [
        'Uber Eats', 'BP Service Station', 'Kmart', 'Starbucks',
        'Returned Headphones', 'Netflix', 'Coles', 'Transperth',
        'Event Cinema', 'Local Restaurant'
    ]
}

df_sample = pd.DataFrame(sample_transactions)
print(df_sample)


# -------------------------
# 1. Data Loading and Cleaning
# -------------------------

def load_and_clean_transaction_data(file_path):
    required_columns = ['Date', 'Amount', 'Category', 'Description']

    try:
        if isinstance(file_path, pd.DataFrame):
            df = file_path.copy()
        else:
            df = pd.read_csv(file_path)

        missing_columns = [col for col in required_columns if col not in df.columns]

        if missing_columns:
            raise ValueError(f"Missing required columns: {missing_columns}")

        df = df[required_columns]

        df['Amount'] = df['Amount'].astype(str).replace('[\\$,]', '', regex=True)
        df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

        df = df.dropna(subset=['Date', 'Amount', 'Category'])

        df['Category'] = df['Category'].astype(str).str.strip()
        df['Description'] = df['Description'].astype(str).str.strip()

        return df

    except Exception as error:
        print(f"Error loading transaction data: {error}")
        return pd.DataFrame()


clean_data = load_and_clean_transaction_data(df_sample)
print(clean_data)



# Test your function
# clean_data = load_and_clean_transaction_data(df_sample)
# print(clean_data)
````

````python
def analyze_spending_patterns(df):
    positive_spending = df[df['Amount'] > 0]
    refunds = df[df['Amount'] < 0]

    total_spending = positive_spending['Amount'].sum()
    total_refunds = refunds['Amount'].sum()
    net_spending = df['Amount'].sum()
    average_transaction = positive_spending['Amount'].mean()

    category_summary = positive_spending.groupby('Category')['Amount'].sum().sort_values(ascending=False)

    if not category_summary.empty:
        highest_category = category_summary.idxmax()
        highest_category_amount = category_summary.max()
    else:
        highest_category = "No spending data available"
        highest_category_amount = 0

    category_percentages = (category_summary / total_spending * 100).round(2)

    return {
        "total_spending": total_spending,
        "total_refunds": total_refunds,
        "net_spending": net_spending,
        "average_transaction": average_transaction,
        "category_summary": category_summary,
        "category_percentages": category_percentages,
        "highest_category": highest_category,
        "highest_category_amount": highest_category_amount
    }


analysis = analyze_spending_patterns(clean_data)
print(analysis)
# Test your function
# analysis = analyze_spending_patterns(clean_data)
# print(analysis)
````

````python
# 🤖 AI Collaboration: Custom Tool Development
# Ask AI to help you create useful financial calculators

from hands_on_ai import agent

def create_savings_calculator_tool(current_savings, monthly_contribution, target_amount):
    if monthly_contribution <= 0:
        return "Monthly contribution must be greater than zero."

    if target_amount <= current_savings:
        return "You have already reached your savings goal."

    remaining_amount = target_amount - current_savings
    months_needed = remaining_amount / monthly_contribution

    return f"You need approximately {months_needed:.1f} months to reach your savings goal."


savings_result = create_savings_calculator_tool(500, 150, 2000)
print(savings_result)


# Register your tool with the agent system
# agent.register_tool("savings_calculator", create_savings_calculator_tool)
````

````python
# 🤖 AI Collaboration: Professional UI Design
# Ask AI to help you create a comprehensive Gradio interface

import gradio as gr

def analyze_uploaded_file(file):
    try:
        data = pd.read_csv(file.name)
        cleaned = load_and_clean_transaction_data(data)
        result = analyze_spending_patterns(cleaned)
        return generate_financial_recommendations(result)
    except Exception as error:
        return f"Error processing file: {error}"


def create_finance_assistant_ui():
    with gr.Blocks() as demo:
        gr.Markdown("# Budget Buddy - Smart Finance Assistant")
        gr.Markdown("Upload CSV transaction data, get spending insights, ask budgeting questions, and calculate savings goals.")

        with gr.Tab("CSV Spending Analysis"):
            file_input = gr.File(label="Upload CSV Transaction File")
            analysis_output = gr.Textbox(label="Financial Analysis Report", lines=12)
            analyze_button = gr.Button("Analyse Spending")
            analyze_button.click(analyze_uploaded_file, inputs=file_input, outputs=analysis_output)

        with gr.Tab("Savings Goal Calculator"):
            current_savings = gr.Number(label="Current Savings")
            monthly_contribution = gr.Number(label="Monthly Contribution")
            target_amount = gr.Number(label="Target Amount")
            savings_output = gr.Textbox(label="Savings Goal Result")
            savings_button = gr.Button("Calculate")
            savings_button.click(
                create_savings_calculator_tool,
                inputs=[current_savings, monthly_contribution, target_amount],
                outputs=savings_output
            )

        with gr.Tab("Budget Buddy Chatbot"):
            question_input = gr.Textbox(label="Ask Budget Buddy a finance question")
            chatbot_output = gr.Textbox(label="Chatbot Response", lines=8)
            chatbot_button = gr.Button("Ask")
            chatbot_button.click(ask_budget_buddy, inputs=question_input, outputs=chatbot_output)

        with gr.Tab("RAG Budgeting Guidance"):
            rag_question = gr.Textbox(label="Ask for a budgeting tip")
            rag_output = gr.Textbox(label="Retrieved Budgeting Advice", lines=6)
            rag_button = gr.Button("Get Advice")
            rag_button.click(simple_finance_rag, inputs=rag_question, outputs=rag_output)

    return demo


demo = create_finance_assistant_ui()
demo.launch()

# Launch your complete application
# demo = create_finance_assistant_ui()
# demo.launch()
````
